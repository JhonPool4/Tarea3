```python
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
# Las imágenes serán normalizadas usando 1/255
datagen = ImageDataGenerator(rescale=1/255)
# Leer imágenes de una carpeta
#   * Carpeta: pregunta2
#   * target_size: todas las imágenes serán escaladas a este tamaño
#   * batch_size: número de imágenes por batch
#   * class_model: binary (solo se predice 2 clases) o categorical (si hay más clases)
datos_train = datagen.flow_from_directory("pregunta3", target_size=(150, 150),
                                          batch_size=10, class_mode='binary')

print("Tamaño de las imágenes:", datos_train.image_shape)
```

```python
# Obtener un batch de las imágenes
Xbatch, ybatch = datos_train.next()
print("Elementos de un batch: ", Xbatch.shape)
# Mostrar las imágenes del batch 
nimag = 10
for i in range(nimag):
    plt.subplot(1, nimag, i+1)
    plt.imshow(Xbatch[i,:])
    plt.title(ybatch[i])
    plt.axis('off')
```
